```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plot of f(x) = x^2
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x, y)
plt.title('Graph of f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

```python
# plot of the normal distribution
mean = 0
std_dev = 1
x = np.linspace(-5,5,100)
# f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}

pdf = (1/(std_dev * (np.sqrt(2*np.pi)))) * (np.exp(-((x - mean)**2)/(2*(std_dev**2))))

plt.plot(x, pdf)
plt.title('Normal Distribution (mean=0, std_dev=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid()
plt.show()
```
